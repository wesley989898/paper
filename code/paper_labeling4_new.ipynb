```python
import pandas as pd
from pandas import DataFrame
from datetime import datetime, timedelta
import dateutil.relativedelta
import re
```

## 計算發佈新聞前一天的收盤價

```python
df = pd.read_excel('stock_data/stockdata_sentence_withdo1.xlsx')
pattern = '%Y/%m/%d'

count = 0

l1 = [] #title
l2 = [] #content
l3 = [] #key
l4 = [] #tm
l6 = [] #code
l7 = [] #start_date
l8 = [] #start_price

#找到股票代號的資料夾
for i, row in df.iterrows():
    try:
        dfs = pd.read_excel(('stock_data/stockprice/{}.xlsx').format(row['code']))
        start_price = 0
        start_date = "None"
    
    
        #往回找新聞前一天。
        if start_price == 0:
            for num in range(1, 15):
                postdate = datetime.strptime(str(row['tm']), pattern)
                fixdate_str = str(postdate + dateutil.relativedelta.relativedelta(days=-int(num))).split(' ')[0]
                fixdate = fixdate_str.replace('-','/')

                for i, rows in dfs.iterrows():
                    if rows['日期'] == fixdate:
                        start_price = rows['收盤']
                        start_date = rows['日期']
                    if not start_price == 0:
                        break
                        
        l1.extend([row['title']])
        l2.extend([row['content']])
        l3.extend([row['cat']])
        l4.extend([row['tm']])
        l6.extend([row['code']])
        l7.extend([start_date])
        l8.extend([start_price])
    
    except:
        #print(row['code'])
        start_price = 0
        start_date = 0

    count = count + 1
    if (count%1000==0):
        print(count)

result = DataFrame({'title': l1, 'content': l2, 'cat':l3, 'tm': l4, 'code': l6, 'start_date': l7, 'start_price': l8})
result
```

## 計算新聞發佈後第n天的開盤價

```python
l1 = [] #title
l2 = [] #content
l3 = [] #key
l4 = [] #tm
l6 = [] #code
l7 = [] #start_date
l8 = [] #start_price
l9 = [] #last_date
l10 = [] #last_price
l11 = [] #labeling
pattern2 = '%Y/%m/%d'

for i, row in result.iterrows():
    try:
        dfs = pd.read_excel(('stock_data/stockprice/{}.xlsx').format(row['code']))
        last_price = 0
        last_date = "None"
    
        for i, rows in dfs.iterrows():
            pg_date_str = datetime.strptime(row['start_date'], pattern2)
            pg_date = str(pg_date_str + dateutil.relativedelta.relativedelta(days=+int(1))).split(' ')[0]
            pg_date = pg_date.replace('-','/')
            if pg_date == rows['日期']:
                last_price = rows['開盤']
                last_date = rows['日期']
            
        if last_price == 0:
            for num in range(1, 15):
                pgdate_str = datetime.strptime(row['start_date'], pattern2)
                pgdate = str(pgdate_str + dateutil.relativedelta.relativedelta(days=+int(num + 1))).split(' ')[0]
                pgdate = pgdate.replace('-','/')
                
                for i, rows in dfs.iterrows():
                    if pgdate == rows['日期']:
                        last_price = rows['開盤']
                        last_date = rows['日期']
                    if not last_price == 0:
                        break
                    
        if ( last_price - row['start_price'] ) > 0:
            labeling = '1'
        elif ( last_price - row['start_price'] ) < 0:
            labeling = '0'
        else:
            labeling = ''
 
        l1.extend([row['title']])
        l2.extend([row['content']])
        l3.extend([row['cat']])
        l4.extend([row['tm']])
        l6.extend([row['code']])
        l7.extend([row['start_date']])
        l8.extend([row['start_price']])                
        l9.extend([last_date])
        l10.extend([last_price])
        l11.extend([labeling])

    except:
        start_price = 0
        start_date = 0
        
result2 = DataFrame({'title': l1, 'content': l2, 'cat':l3, 'tm': l4, 'code': l6, 'start_date': l7, 'start_price': l8, 'last_date': l9, 'last_price': l10, 'labeling': l11})
result2.to_excel('stock_data/stockdata_finish_new1.xlsx')
```
