```python
import pandas as pd
from pandas import DataFrame
```

```python
df_negative = pd.read_excel('dictionary/negative_total.xlsx')
df_positive = pd.read_excel('dictionary/positive_total.xlsx')
df_dev = pd.read_excel('stock_data/stockdata_withdo_new_clean1.xlsx')
```

```python
positive = set([x.strip() for x in df_positive['sentiment']])
negative = set([x.strip() for x in df_negative['sentiment']])
total = []

for i, row in df_dev.iterrows():
    sentiment = []
    list1 = row['content'].split()
    
    # 每個詞作比對，word要加[]是因為超過兩個字，set變成list格式
    for word in list1:
        if set([word]) & positive:
            sentiment.append(1)
        elif set([word]) & negative:
            sentiment.append(2)
        else:
            sentiment.append(0)
    total.append(sentiment)
    
    if i%1000 == 0:
        print('Now:' + str(i))

df_dev['sentiment'] = total
df_dev.to_excel('stock_data/stockdata_sentiment_new_withdo1.xlsx')
```
