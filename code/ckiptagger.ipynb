```python
from ckiptagger import data_utils, construct_dictionary, WS, POS, NER
import pandas as pd
from pandas import DataFrame
import xlrd
```

```python
#建立自行詞典
df = pd.read_excel('stock_company.xlsx')

values = []  #公司名稱
corpus = {}  #公司名稱 & 權重 ex:'三商': 2.0
key = 2 #權重

#將公司名稱裝入list
for i, row in df.iterrows():
    values += [ row['company'] ]

#建立dict
for i in values:
    corpus[i] = 2

dictionary = construct_dictionary(corpus)
dictionary
```

```python
#進行斷詞

df = pd.read_excel("stock_data/stockdata1.xlsx") 
l1 = []
l2 = []
l3 = []
stopWords=[",", "." ,"、", ":", "?", "!", ";", "，", "◆", "■", "《", "》", "(", ")", "：" , "'", '"', "&emsp;", "；", " ", "（", "）", "」", "「", "\r", "...", "★", "◇", "【", "】" ,"、","『","／"]

count = 0
for i, row in df.iterrows():
    sentence = ''
    ws = WS("./data")
    
    title = row['title']
    l1.extend([title])

    #斷詞, 導入自行建構字典
    try:
        content = list(filter(lambda a: a not in stopWords and a != '\n', row['content'])) #去除停用字
        for i in content: sentence += i #斷詞結果合併
    except:
        content = row['content']
    
    ws_results = ws([sentence], recommend_dictionary = dictionary) #斷詞
    l2.extend(ws_results) 

    time = row['tm']
    l3.extend([time])
    count = count + 1
    if (count%1000 == 0):
        print(count)

df = DataFrame({'title': l1, 'content': l2, 'tm': l3})
df.to_excel('stock_data/stockdata_cut_withdo1.xlsx')
```
